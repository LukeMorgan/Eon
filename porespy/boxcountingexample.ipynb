## PORESPY TUTORIAL 3: Measuring Fractal Dimension by Box Counting.

Mandelbrot introduced the idea of fractal dimensions to explain the self-similarity of a pattern. It is defined as a ratio of the change in detail to the change in scale. It quantifies the complexity of a fractal pattern in a neat index.

One way to determine the fractal dimension of an image is the box counting method. Boxes of various sizes are laid over the image in a fixed grid pattern. The number of boxes that span the edge of the pattern (part is in region 1, part is in region 2) are tallied as a fucntion of box size. This count is then used to calculate final dimension.

### Mathematical Definition

The relationship of a pattern's fractal dimension $F$ and it's measuring element can be expressed as:

$N \propto \frac{1}{D^F}$

where

$F = \displaystyle{\lim_{D \to 0}}\frac{-log(N(D))}{log(D)}$ 

and 

$N$ is number of boxes with side length $D$ that span an edge.

### Example

A Sierpinksi carpet has a known fractal dimension of 1.8928. Performing the box counting method found its fractal dimension as approximately 1.8-1.9.

```python
import matplotlib.pyplot as plt
import porespy as ps
ps.visualization.set_mpl_style()
```

```python
im = ps.generators.sierpinski_foam(4,5)
plt.imshow(im);
```

Finally, apply the box count function and visualize. 

```python
data = ps.metrics.boxcount(im)
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(8,4))
ax1.set_yscale('log')
ax1.set_xscale('log')
ax1.set_xlabel('box edge length')
ax1.set_ylabel('no. boxes spanning both phases')
ax2.set_xlabel('box edge length')
ax2.set_ylabel('slope')
ax2.set_xscale('log')
ax1.plot(data.size, data.count,'-o')
ax2.plot(data.size, data.slope,'-o');
```

On the right image, between 10 and 100 length we can see a flat slope with a value of approximately 1.9. Beyond a box edge length of 100, the analysis becomes impacted by finite image size, so the results begin to diverge.
